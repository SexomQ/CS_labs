```python
# task 1.1
def menu():
    print("Choose encryption/decryption (1/2)")
    print("Exit (0)")
    choice = None
    while choice != 0:
        choice = int(input("Your choice here: "))
        if choice == 1:
            print("You chose encryption!")
            print(encryption())
        elif choice == 2:
            print("You chose decryption!")
            print(decryption())
        elif choice == 0:
            exit()
            break
        else:
            print("Not valid input! Dont be dumb...")

def exit():
    print("Exiting the program...")
    time.sleep(3)
    print("Finished!")   

def encryption():
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    Alphabet = [letter.upper() for letter in alphabet]

    valid_key = False
    while valid_key == False:
        key = int(input("Introduce the key: "))

        if key > 0 and key < 24:
            valid_key = True

    print("The key:", key)
    
    valid_word = False
    while valid_word == False:
        message = input("Introduce the message: ")
        message = message.replace(" ", "")
        valid_word = message.isalpha()

        if not valid_word:
            print("Not valid message! Dont be dumb... choose letters!")
    
    print("The message is:", message)

    trans_message = message.upper()
    
    letters_orig = [a for a in trans_message]
    letters_out = letters_orig.copy()
    for i in range(len(letters_orig)):
        alpha_index_crypt = (Alphabet.index(letters_orig[i]) + key) % len(alphabet)
        letters_out[i] = Alphabet[alpha_index_crypt]
        
    output = ''.join(letters_out)
    output = output.lower()
    
    return f"The encrypted message is {output}"

def decryption():
    alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']
    Alphabet = [letter.upper() for letter in alphabet]

    valid_key = False
    while valid_key == False:
        key = int(input("Introduce the key: "))

        if key > 0 and key < 24:
            valid_key = True

    print("The key:", key)
    
    valid_word = False
    while valid_word == False:
        message = input("Introduce the cryptograme: ")
        message = message.replace(" ", "")
        valid_word = message.isalpha()

        if not valid_word:
            print("Not valid message! Dont be dumb... choose letters!")
    
    print("The message is:", message)

    trans_message = message.upper()

    letters_orig = [a for a in trans_message]
    letters_out = letters_orig.copy()
    for i in range(len(letters_orig)):
        alpha_index_crypt = (Alphabet.index(letters_orig[i]) - key) % len(alphabet)
        letters_out[i] = Alphabet[alpha_index_crypt]
        
    output = ''.join(letters_out)
    output = output.lower()
    
    return f"The decrypted message is {output}"

menu()     
```
